```python
def make_genre_matrix(df, target_column, desired_row_index):
    
    topic_list = np.unique(np.concatenate(df[target_column].values))
    num_obs = len(df[target_column])
    num_topics = len(topic_list)
    results_array = np.empty((num_obs, num_topics))
    for row_idx in range(num_obs): 
        results_array[row_idx]= np.isin(topic_list, df.iloc[row_idx, 2])
    
    results_df = pd.DataFrame(results_array, columns=topic_list, index = df[desired_row_index])
    return results_df

# genres_df = make_genre_matrix(movies, 'genres', 'movieId')
# genres_df.head()
```

```python
def find_the_most_similar(movie_idx, num_to_find, sim_mtx, movie_mtx):
    '''
    idx: integer type, index of the movie we care about
    num_to_find: integer type, number of recommendations requested
    sim_mtx: similarity matrix of all movies compared to each other
    movie_mtx: original matrix of movies with names
    '''
    row = sim_mtx[movie_idx]
    original_movie = movie_mtx[movie_mtx['movieId']==movie_idx]
    top_5_idx = np.argsort(row)[-1*(num_to_find+1):-1]
    top_5_ids = sim_mtx.columns[top_5_idx]
    
    top_movies = movies[np.isin(movies['movieId'],top_5_ids.values)]
    return original_movie, top_movies
```
